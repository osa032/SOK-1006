### Estimer Z, a, b og forklar nye hva hvert tall representerer. Forklar i detalj hva du gjr.

En Cobb-Douglas-produksjonsfunksjon er en spesifikk type produksjonsfunksjon som brukes i konomi for  beskrive forholdet mellom innsatsfaktorer (input) og produksjon (output).

Den kan brukes til  modellere produksjonsprosesser i ulike industrier. For eksempel, i en fabrikk kan output (produksjon) vre en funksjon av antall arbeidere (arbeidskraft) og mengden maskineri (kapital). Ved  bruke historiske data om output, arbeidskraft og kapital, kan vi estimere verdiene for \(Z\), \(a\) og \(b\) for  forst hvordan endringer i innsatsfaktorer vil pvirke produksjonen.

***Fordeler og begrensninger***

**Fordeler:**
- Gir innsikt i hvordan ulike innsatsfaktorer pvirker produksjonen.

**Begrensninger:**
- Antar konstant skalautbytte over tid og teknologi.
- Forutsetter perfekte markeder og fravr av eksternaliteter.


Forklaring av parameterne i en Cobb-Douglas-funksjon med formen:

$$ Y = Z \cdot N^a \cdot K^b $$

Hvor:
- $ Y $ er produksjonsnivet (output).
- $ Z $ er en konstant som representerer total faktorproduktivitet (TFP), ogs kjent som teknologi eller effektivitet. Dette er en skalar som pvirker hvor effektivt inputs (N og K) blir omdannet til output.
- $ N $ er mengden arbeidskraft som brukes i produksjonen.
- $ K $ er mengden kapital som brukes i produksjonen.
- $ a $ er outputelastisiteten til arbeidskraft, som indikerer prosentvis endring i output nr arbeidskraften ker med 1%, gitt at alle andre faktorer er konstante.
- $ b $ er outputelastisiteten til kapital, som indikerer prosentvis endring i output nr kapitalen ker med 1%, gitt at alle andre faktorer er konstante.

For  estimere parameterne $Z$, $a$ og $b$ i Cobb-Douglas-produksjonsfunksjonen med formen $Y = Z \cdot N^a \cdot K^b$ tar jeg disse stegene:

1. **Logaritmisk transformasjon**:
   Transformerer Cobb-Douglas-funksjonen til en liner form ved bruk av logaritmer for  kunne bruke liner regresjon til  estimere parametrene.
   
   Den logaritmiske formen av Cobb-Douglas-funksjonen er:

   $$log(Y) = \log(Z) + a \cdot \log(N) + b \cdot \log(K)$$

   Hvor $log(Y)$ er $y$, $log(N)$ er $n$, og $log(K)$ er $k$. Da blir ligningen:

   
   $$y = \log(Z) + a \cdot n + b \cdot k$$
   

2. **Data**:
   Av dataene fra df sier vi at:

- $Y$ er totalproduksjonen (Produksjonsverdi),
- $N$ er lnnskostnadene (Lnnskostnader),
- $K$ er bruttoinvesteringene (Bruttoinvesteringer),


3. **Liner regresjon**:
   Etter  ha konvertert likningen til en liner likning, og funnet frem tallene jeg skal bruke kjrer jeg en multippel liner regresjonsanalyse for  estimere koeffisientene. Koeffisienten for konstantleddet vil gi $log(Z)$, og koeffisientene for $n$ og $k$ vil gi henholdsvis $a$ og $b$.

### Forklar skalautbytte, og hvordan man avgjres skalautbytte for en CobbDouglas funksjon

Skalautbytte refererer til hvordan endringen i mengden av alle innsatsfaktorer pvirker mengden av output. For en Cobb-Douglas-produksjonsfunksjon avgjr man typen skalautbytte ved  summere eksponentene $a$ og $b$:

- Hvis $a + b = 1$, er det konstant skalautbytte. Dette betyr at en proporsjonal kning i alle innsatsfaktorer frer til en like stor prosentvis kning i output.
- Hvis $a + b > 1$, er det kende skalautbytte. Dette betyr at en proporsjonal kning i alle innsatsfaktorer frer til en strre prosentvis kning i output.
- Hvis $a + b < 1$, er det avtakende skalautbytte. Dette betyr at en proporsjonal kning i alle innsatsfaktorer frer til en mindre prosentvis kning i output.

### Hva er skalaegenskapen for produktfunksjonen? Er det intuitivt riktig ifht nring?

For  finne skalaegenskapen til denne funksjonen, summerer jeg verdiene for $a$ og $b$:
$$a + b = 0.81 + 0.15 = 0.96$$

Siden $a + b < 1$, er det avtakende skalautbytte. Dette betyr at en proporsjonal kning i bde arbeidskraft (N) og kapital (K) vil fre til en mindre enn proporsjonal kning i output.

Grensekostnad eller marginalkostnad er kostnaden ved  produsere en ytterligere enhet av en vare eller en tjeneste.

Gjennomsnittskostnad er de totale kostnadene (grensekost + fc) delt p antall enheter produsert

Gjennomsnittlig variable kostnader er gjennomsnittelige kostnader uten FC

Vi vet fra forrige deloppgave at uttrykket for arbeidskraft $L$ som en funksjon av produksjon $y$ er:

$$ L = \left( \frac{y}{5.683} \right)^{\frac{1}{0.81}} $$

1. Finn grensekostnad (MC)

Grensekostnadene (MC) er den deriverte av de totale kostnadene (TC) med hensyn til produksjonsnivet $y$:

$$MC(y) = \frac{dTC(y)}{dy} $$

Beregner den deriverte:
$$ VC(y) = 287 \cdot \left( \frac{y}{5.683} \right)^{\frac{1}{0.81}}$$

$$VC(y) = 287 \cdot \left( \frac{y}{5.683} \right)^{1.2346} $$
  $$(siden \ ( \frac{1}{0.81} \approx 1.2346 ))$$

Deriverer s VC(y):
$$MC(y) = \frac{d}{dy} \left[ 287 \cdot \left( \frac{y}{5.683} \right)^{1.2346} \right] $$

$$ MC(y) = 287 \cdot 1.2346 \cdot \left( \frac{y}{5.683} \right)^{0.2346} \cdot \frac{1}{5.683}$$

$$ MC(y) = 287 \cdot 1.2346 \cdot 5.683^{-1.2346} \cdot y^{0.2346} $$

2. Finn gjennomsnittskostnad (AC)

Gjennomsnittskostnadene (AC) er de totale kostnadene (TC) delt p mengde produsert $y$:

$$ AC(y) = \frac{TC(y)}{y} $$

$$ AC(y) = \frac{287 \cdot \left( \frac{y}{5.683} \right)^{1.2346} + 2,519,000}{y} $$

Gjennomsnittlig variable kostnader (AVC):
$$ AVC(y) = \frac{VC(y)}{y} $$

$$ AVC(y) = \frac{287 \cdot \left( \frac{y}{5.683} \right)^{1.2346}}{y} $$

\[ AC(y) = \frac{287 \cdot \left( \frac{y}{5.683} \right)^{1.2346} + 2,519,000}{y} \]

- **Gjennomsnittlig variable kostnader (AVC)**:
\[ AVC(y) = \frac{287 \cdot \left( \frac{y}{5.683} \right)^{1.2346}}{y} \]

### Tegn bedriftens tilbudskurve i en figur og forklar denne nye

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Define the variablesdefinerer variabler
y = sp.symbols('y')
Z = 4.87
a = 0.81
b = 0.15
K0 = 2.519  # i millioern NOK
w = 287  # NOK per time

# Produksjonsfunksjon og arbeidskraft (L)
L_expr = (y / (Z * K0**b))**(1 / a)

# Variable kostnader (VC), FC
VC_expr = w * L_expr
FC = K0 * 1e6  # i NOK
TC_expr = VC_expr + FC

# Marginalkostnad (MC), gj.kos (AC) og gj.gj.kost(AVC) funksjoner
MC_expr = sp.diff(TC_expr, y)
AC_expr = TC_expr / y
AVC_expr = VC_expr / y

# Konvertere uttrykkene til numeriske funksjoner, bruker `sp.lambdify` for  konvertere de symbolske uttrykkene til numeriske funksjoner
MC_func = sp.lambdify(y, MC_expr, 'numpy')
AVC_func = sp.lambdify(y, AVC_expr, 'numpy')

# Generere data for plotting
y_vals = np.linspace(1, 10000, 500)  # Produksjonsniver

MC_vals = MC_func(y_vals)
AVC_vals = AVC_func(y_vals)

# Plotter tilbudskurven
plt.figure(figsize=(10, 6))

# MC
plt.plot(y_vals, MC_vals, label='Grensekostnad (MC)')

# AVC
plt.plot(y_vals, AVC_vals, label='Gjennomsnittelige variable kostnader (AVC)')

# Finn tilbud (delen av MC som er over AVC)
supply_curve_y = y_vals[MC_vals > AVC_vals]
supply_curve_MC = MC_vals[MC_vals > AVC_vals]

plt.plot(supply_curve_y, supply_curve_MC, label='Tilbudskurve', linestyle='--', color='red')

plt.xlabel('Kvantum (y)')
plt.ylabel('Pris (NOK)')
plt.title('Tilbudskurve')
plt.legend()
plt.grid(True)
plt.show()
```

Tilbudskurven til en bedrift er p kort sikt den delen av marginalkostnadskurven (MC) som ligger over den gjennomsnittlige variable kostnadskurven (AVC). Dette er fordi at bedriften bare vil produsere dersom prisen dekker de variable kostnadene.

### Anta at lnnsoppgjret gir arbeiderne i en lnnskning p 5% og vis hvordan den pvirker bedriftens tilbud, forklar intuisjonen bak endringen

1. Oppdatere lnnskostnaden
Den opprinnelige lnnen per time var $ w = 287 $ NOK. Med en lnnskning p 5%, blir den nye lnnen:
$$ w_{\text{ny}} = w \times 1.05 $$

$$ w_{\text{ny}} = 287 \times 1.05 $$

$$ w_{\text{ny}} = 301.35 \, \text{NOK per time} $$

2. Oppdatere kostnadsfunksjonene

- Variable kostnader (VC)
De variable kostnadene med den nye lnnen er:
$$ VC(y) = w_{\text{ny}} \cdot L $$

$$ VC(y) = 301.35 \cdot \left( \frac{y}{5.683} \right)^{\frac{1}{0.81}} $$

- Marginalkostnad (MC)
Marginalkostnaden med den nye lnnen er den deriverte av de totale kostnadene (TC) med hensyn til produksjon $y$:

$$ MC(y) = \frac{d}{dy} \left(301.35 \cdot \left( \frac{y}{5.683} \right)^{\frac{1}{0.81}} + FC \right) $$

$$ MC(y) = 301.35 \cdot 1.2346 \cdot \left( \frac{y}{5.683} \right)^{0.2346} \cdot \frac{1}{5.683} $$

$$ MC(y) = 301.35 \cdot 1.2346 \cdot 5.683^{-1.2346} \cdot y^{0.2346} $$

-  Gjennomsnittlig variable kostnader (AVC)
Gjennomsnittlig variable kostnader med den nye lnnen er:
$$ AVC(y) = \frac{VC(y)}{y} $$

$$ AVC(y) = \frac{301.35 \cdot \left( \frac{y}{5.683} \right)^{1.2346}}{y} $$

3. Oppdatere og plotte kostnadskurvene


```python

import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Definer variabler og parametere
y = sp.symbols('y')
Z = 4.87
a = 0.81
b = 0.15
K0 = 2.519  # i million NOK
w = 287  # NOK per time
w_new = w * 1.05  # Ny lnn etter 5% lnnskning

# Produksjonsfunksjon og arbeidskraft (L)
L_expr = (y / (Z * K0**b))**(1 / a)

# Variable kostnader (VC) - fr og etter lnnskning
VC_expr = w * L_expr
VC_new_expr = w_new * L_expr

# Marginalkostnad (MC) - fr og etter lnnskning
MC_expr = sp.diff(VC_expr, y)
MC_new_expr = sp.diff(VC_new_expr, y)

# Gjennomsnittlig variable kostnader (AVC) - fr og etter lnnskning
AVC_expr = VC_expr / y
AVC_new_expr = VC_new_expr / y

# Konvertere uttrykkene til numeriske funksjoner, bruker `sp.lambdify` for  konvertere de symbolske uttrykkene til numeriske funksjoner
MC_func = sp.lambdify(y, MC_expr, 'numpy')
MC_new_func = sp.lambdify(y, MC_new_expr, 'numpy')
AVC_func = sp.lambdify(y, AVC_expr, 'numpy')
AVC_new_func = sp.lambdify(y, AVC_new_expr, 'numpy')

# Generere data for plotting
y_vals = np.linspace(1, 100, 500)  # Produksjonsniver

MC_vals = MC_func(y_vals)
MC_new_vals = MC_new_func(y_vals)
AVC_vals = AVC_func(y_vals)
AVC_new_vals = AVC_new_func(y_vals)

# Plotter
plt.figure(figsize=(10, 6))

# Plot marginalkostnadskurvene
plt.plot(y_vals, MC_vals, label='Marginalkostnad (MC) - fr lnnskning')
plt.plot(y_vals, MC_new_vals, label='Marginalkostnad (MC) - etter lnnskning', linestyle='--')

# Plot gjennomsnittlig variable kostnadskurvene
plt.plot(y_vals, AVC_vals, label='Gjennomsnittelige variable kost (AVC) - fr lnnskning')
plt.plot(y_vals, AVC_new_vals, label='Gjennomsnittelige variable kost (AVC) - etter lnnskning', linestyle='--')

# Identifisere og highlight tilbudskurven (delen av MC over AVC)
supply_curve_y_before = y_vals[MC_vals > AVC_vals]
supply_curve_MC_before = MC_vals[MC_vals > AVC_vals]
supply_curve_y_after = y_vals[MC_new_vals > AVC_new_vals]
supply_curve_MC_after = MC_new_vals[MC_new_vals > AVC_new_vals]

plt.plot(supply_curve_y_before, supply_curve_MC_before, label='Tilbudskurve - fr lnnskning', linestyle=':', color='green')
plt.plot(supply_curve_y_after, supply_curve_MC_after, label='Tilbudskurve - etter lnnskning', linestyle='-.', color='red')

plt.xlabel('Produksjon (y)')
plt.ylabel('Pris (NOK)')
plt.title('Tilbudskurve fr og etter lnnskning')
plt.legend()
plt.grid(True)
plt.show()
```



4. Intuisjonen bak endringen

- **kte kostnader**: En lnnskning p 5% ker de variable kostnadene fordi arbeidskraft blir dyrere.
- **Marginalkostnadskurve**: MC-kurven skifter oppover fordi det koster mer  produsere hver ekstra enhet.
- **Tilbudskurve**: Tilbudskurven, som er delen av MC-kurven over AVC, skifter ogs oppover. Dette betyr at for en gitt produksjonsmengde, vil bedriften kreve en hyere pris for  dekke de kte kostnadene.
- **Markedseffekter**: P markedet kan dette fre til hyere priser og redusert tilbud.
